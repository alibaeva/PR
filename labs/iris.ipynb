```python
import pandas as pd
import numpy as np
from ipywidgets import interact_manual
@interact_manual
def get_input(length='5', width ='3', petallength = '1.5', petalwidth = '0.1'):
    defined = 0
    if float(petalwidth) < 1:
        print('0 - setosa')
        defined = 1
    elif float(petalwidth) >=1 and float(petalwidth) <=1.6:
        print('1 - versicolor')
        defined = 1
    elif float(petalwidth) >=1.8:
        print('2 - virginica')
    if defined == 0:
        if float(length) >= 5.7 and float(width) >= 2.5 and petallength >= 4.5 and petalwidth >= 1.6:
            print('2 - virginica')
        else: 
            print('1 - versicolor')
```
